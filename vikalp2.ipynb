```python
import pandas as pd


url = 'https://raw.githubusercontent.com/taylorgaines/songs/main/Khalid.csv'
data = pd.read_csv(url)

data.head()


data.to_csv('/content/Khalid.csv', index=False)
data.head()
```

```python
year_column = 'Year'
year_frequency = data[year_column].value_counts()
year_frequency.sort_values()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.bar(year_frequency.index.astype(int),year_frequency.values, color='lightblue')

plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.title('Song Frequency by Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
Album_column = 'Album'
Album_frequency = data[Album_column].value_counts()
Album_frequency.sort_values()
```

```python

plt.figure(figsize=(10, 6))
plt.bar(Album_frequency.index, Album_frequency.values, color='lightblue')

plt.xlabel('Album')
plt.ylabel('Number of Songs')
plt.title('Number of Songs per Album')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
from textblob import TextBlob
data['sentiment'] = data['Lyric'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
data[['Title', 'sentiment']].head(10)
print(data[['Title', 'sentiment']])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(data['sentiment'], bins=10, color='skyblue', edgecolor='black')
plt.title('Sentiment Distribution of Khalid Songs')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Number of Songs')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
from collections import Counter
all_words = ' '.join(data['Lyric'].dropna()).lower().split()
common_words = Counter(all_words).most_common(20)
print(common_words)
```

```python
import matplotlib.pyplot as plt

# Unpack the data
words, frequencies = zip(*common_words)

x = range(1, len(words) + 1)

plt.figure(figsize=(12, 6))
plt.plot(x, frequencies, marker='o', color='skyblue', linestyle='-')
plt.yscale('log')
plt.title('Word Frequency Line Graph (Log Scale)')
plt.xlabel('Word Rank')
plt.ylabel('Log Frequency')
plt.xticks(ticks=x, labels=words, rotation=45)
plt.tight_layout()
plt.show()
```
